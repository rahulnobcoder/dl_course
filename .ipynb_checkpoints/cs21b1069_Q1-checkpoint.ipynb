```python
#cs21b1069
#Rahul.k
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('iris.csv')
data
```

```python
data=data[data['Species']!='Iris-virginica']
```

```python
data['Species']=data['Species'].astype('category')#encoding two species of flowers as 0 and 1 using encoding below
data['Species']=data['Species'].cat.codes
data
```

```python
X=data.drop(columns=['Species','Id'])
X['bias']=1
y=data['Species']
X
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
def step_function(x):
    arr=[]
    for i in x:
        if i>=0:
            arr.append(1)
        else:
            arr.append(0)
    return np.array(arr)
```

```python
w=np.array([0]*5)     #weight vector
a=0.0001    #learning rate
iterations=100000    #number of iterations
```

```python
def gradient_descent(w,alpha):
    yprediction=np.matmul(xtrain,np.transpose(w))
    yprediction=step_function(yprediction)
    error=(yprediction-np.array(ytrain))
    derivative=[0]*5
    for i in range(len(error)):
        derivative+=xtrain.iloc[i]*error[i]
    w=w-alpha*derivative
    return np.array(w)
```

```python
count=0
for i in range(iterations):
    w_=gradient_descent(w,a)
    compare =  w==w_
    count+=1
    if compare.all(): #to check if the derivative of error fucntion has become zero or not(convergence)
        break
    w=w_
    print(f"iteration {i+1}:")
    print("Weight vector w =",w)
    yprediction=np.matmul(xtest,w)
    yprediction=step_function(yprediction)
    accuracy=100*sum(ytest==yprediction)/len(yprediction)
    print("accuracy :",accuracy)
print(f"error is minimized at iteration {count-1}")
print(f"weight vector after iteration {count} remains same as it converges")
```
